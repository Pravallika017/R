```python
import pandas as pd
```

```python
df=pd.read_csv("/content/data_ted_talks.csv")
```

```python
df
```

```python
# prompt: linear regression graph

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming 'views' and 'comments' are the columns you want to use for regression
X = df['views'].values.reshape(-1, 1)  # Independent variable (views)
y = df['comments'].values  # Dependent variable (comments)

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the comments based on the views
y_pred = model.predict(X)

# Plot the data points and the regression line
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Views')
plt.ylabel('Comments')
plt.title('Linear Regression: Views vs. Comments')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Remove rows with NaN values in 'views' or 'comments'
df = df.dropna(subset=['views', 'comments'])

# Extract features and target variable
X = df['views'].values.reshape(-1, 1)  # Independent variable (views)
y = df['comments'].values  # Dependent variable (comments)

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the comments based on the views
y_pred = model.predict(X)

# Plot the data points and the regression line
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Views')
plt.ylabel('Comments')
plt.title('Linear Regression: Views vs. Comments')
plt.legend()
plt.show()
```
